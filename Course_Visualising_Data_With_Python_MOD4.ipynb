Let's continue adding some new teaks to the plot:

```python
plt.figure(figsize=(15,10))

#increase size of all text
sns.set(font_scale=1.5)

#change background to white
sns.set_style('ticks')

#increases the size of the marker shape
ax=sns.regplot(x='year',y='total',data=df_tot,color='green',marker='+',scatter_kws={'s':200})

#adding x- and y-labels
ax.set(xlabel='Year',ylabel='Total Immigration') 
#adding title
ax.set_title('Total Immigration to Canadafrom 1980- 2013')
```

```python
plt.figure(figsize=(15,10))

#increase size of all text
sns.set(font_scale=1.5)

#change background to white but with gridlines
sns.set_style('whitegrid')

#increases the size of the marker shape
ax=sns.regplot(x='year',y='total',data=df_tot,color='green',marker='+',scatter_kws={'s':200})

#adding x- and y-labels
ax.set(xlabel='Year',ylabel='Total Immigration') 
#adding title
ax.set_title('Total Immigration to Canadafrom 1980- 2013')
```
